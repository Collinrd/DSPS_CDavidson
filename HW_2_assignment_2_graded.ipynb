# Normal

```python
distributions = ['gaus']
```

```python
mymean = 100
```

```python
import numpy as np
np.random.seed(777)

mydists = {}

mydists['gaus'] = {}

mydists['gaus'][100] = np.random.normal(mymean, size=100)

pl.hist(mydists['gaus'][100], bins = 30)
pl.ylabel('N')
pl.xlabel('x');
```

Normal distribution of a sample of size 100 for a normal distribution

```python
print("gaus mean: %.2f, standard deviation: %.2f"%(mydists['gaus'][100].mean(), mydists['gaus'][100].std()))
```

```python
mysize = (2000 / (np.array(range(1, 100)))).astype(int)

print ("shape of the 'size' list that contains the size of each distribution", mysize.shape)
```

```python
pl.plot(mysize, '.')
pl.xlabel("index")
pl.ylabel("size of the distributions to be created");
#delete??
```

```python
for n in mysize:
    mydists['gaus'][n] = np.random.normal(mymean, size = n) 

mydists['gaus']['means'] = {}

axchisq_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in mydists['gaus']:
    if not type(nn) == str: 
        mydists['gaus']['means'][nn] = mydists['gaus'][nn].mean()
        axchisq_mu_n.plot(nn, mydists['gaus']['means'][nn], 'o')
        axchisq_mu_n.set_xlabel('sample size', fontsize=18)
        axchisq_mu_n.set_ylabel('sample mean', fontsize=18)
axchisq_mu_n.set_title('Normal', fontsize=18)
axchisq_mu_n.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
axchisq_mu_n.text(axchisq_mu_n.get_xlim()[1] * .6, 100, "expected mean", 
                  va="bottom", fontsize=20)
```

The distribution of sample means here forms similar to a normal distribution showing that with more and more samples the averages will tend toward the expected which is consistant with the law of large numbers.

```python
allmeans = list(mydists['gaus']['means'].values())

pl.figure(figsize=(10, 10))
pl.hist(allmeans,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

The distribution of all sample means, which forms a normal distribution.

# The Central Limit Theorem

The distributions for the means of all samples for the normal, poisson, binomial, chi-squared, and cauchy distributions all form a histogram with an approximately normal distribution. The Central Limit Theorem states that the properly normalized sum of independent random variables tends toward a normal distribution even if the original variables themselves are not normally distributed. This is consistant with the results showing that the distributions of all the sample means from different kinds of distributions tends toward normal.
